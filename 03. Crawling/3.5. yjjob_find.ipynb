```python
# import 
import requests
from bs4 import BeautifulSoup

# page 가져와서 파싱
page = requests.get("http://www.yjjob.or.kr/p/?j=23")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
table = soup.find("table","tbl-c")
rows = table.find_all("td","lh_140 v_top relative")

list_all = []
list = []

for row in rows :
    title = row.select_one("span.font14.bold").text # 과정명
    term = [row.select_one("p.tpad3>em.blue").text,row.select_one("p.tpad3>em.red").text]# 교육기간
    code = row.select("p.tpad5>span.red.bold")
        
    # 코드 제외 p.tpad5
    time = row.select("p.tpad5 span:not(.red)")[0].text.strip()
    price = row.select("p.tpad5 span:not(.red)")[1].text.strip() 
    cost = row.select("p.tpad5 span:not(.red)")[2].text.strip()    

    
    print("과정명: ", title)
    print("교육 기간: ", term[0]+" ~ "+ term[1])
    print("교육시간/요일: ", time)
    print("수강료: ", price)
    print("훈련수당: ", cost)
    print("\n--------------------------\n")

```
